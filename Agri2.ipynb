```python
#Finding Seasonality
#A repeating pattern within each year is known as seasonal variation.
#Types:Time of Day.Daily.Weekly.Monthly.Yearly.
#The model of seasonality can be removed from the time series. This process is called Seasonal Adjustment, or Deseasonalizing.
```
